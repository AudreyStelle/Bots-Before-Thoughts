## Final Product Description

This infographic explores where AI-related jobs are projected to grow and how that growth differs across industries in terms of pay, remote work flexibility, and automation risk. Using industry-level summaries, the visualization highlights that AI job growth is concentrated in a small number of sectors and that high opportunity often comes with important tradeoffs.

Together, the plots show that industries with strong projected growth and higher salaries do not necessarily face lower automation risk, and that flexibility varies widely across sectors. The infographic is designed to communicate these patterns clearly to a general audience while grounding all insights in the underlying data.
